```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score ,classification_report , confusion_matrix 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_classification # data generator
```

```python
x,y= make_classification(n_samples=500 , n_features=2 , n_informative=2 , n_redundant=0 , n_repeated=0 , n_classes=2 , flip_y=0.01 , random_state=42)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3 , random_state=42)
```

```python
param_grid = {'n_neighbors': range(1, 11)}  

knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=5)
grid_search.fit(x_train, y_train)


best_k = grid_search.best_estimator_
print("Best Neighbours for k", grid_search.best_params_)
```

```python
KNN_best = KNeighborsClassifier(n_neighbors= 3)
KNN_best.fit(x_train,y_train) # train model
y_pred = KNN_best.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print( accuracy)
```

```python
k_values = range(1,11)
accuracies = []
```

```python
for k in k_values :
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    accuracies.append(accuracy_score(y_test,y_pred))
```

```python
plt.figure(figsize=(8,5))
plt.plot(k_values, accuracies, marker='o', linestyle='-', color='b', markerfacecolor='red' , markersize=8)
plt.title("Accuracy vs. Number of Neighbors (k)")
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.grid(True)
plt.xticks(k_values)  
plt.show()
```
